```python
import pandas as pd
import sqlite3

tmdb = pd.read_csv('zippedData/tmdb.movies.csv.gz')
rt_reviews = pd.read_csv('zippedData/rt.reviews.tsv.gz', sep='\t', encoding= 'unicode_escape')
rt_movie_info = pd.read_csv('zippedData/rt.movie_info.tsv.gz', sep='\t')
bom_movie_gross = pd.read_csv('zippedData/bom.movie_gross.csv.gz')
tn_movie_budgets = pd.read_csv('zippedData/tn.movie_budgets.csv.gz')
conn = sqlite3.connect('zippedData/im.db/im.db')
conn.close()


tmdb_genres={
'Action':          28,
'Adventure':       12,
'Animation':       16,
'Comedy':          35,
'Crime':           80,
'Documentary':     99,
'Drama':           18,
'Family':          10751,
'Fantasy':         14,
'History':         36,
'Horror':          27,
'Music':           10402,
'Mystery':        9648,
'Romance':         10749,
'Science Fiction': 878,
'TV Movie':        10770,
'Thriller':        53,
'War':             10752,
'Western':         37}

tmdb_genres_swap = {y:x for x,y in tmdb_genres.items()}
tmdb['genre_ids'] = tmdb['genre_ids'].apply(lambda x: [tmdb_genres_swap.get(int(id), 'X') for id in x.strip('[]').split(',') if len(id) != 0])
tmdb

```
